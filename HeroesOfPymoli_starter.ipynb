```python
# Dependencies and Setup
import pandas as pd

# File to Load (Remember to Change These)
file_to_load = "Resources/purchase_data.csv"

# Read Purchasing File and store into Pandas data frame
purchase_data = pd.read_csv(file_to_load)
purchase_data.head()
```

## Player Count

* Display the total number of players


```python
#Display the total numbers of players
Players= len(purchase_data["SN"].value_counts())
Player_Count_df=pd.DataFrame({"Total Players": [Players]})
Player_Count_df
```

## Purchasing Analysis (Total)

* Run basic calculations to obtain number of unique items, average price, etc.


* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame


```python
# Purchasing Analysis Total 
Number_unique_items = len(purchase_data["Item ID"].unique())
avg_price = purchase_data["Price"].mean()
number_purchases = len(purchase_data)
total_revenue = purchase_data["Price"].sum()

# Create df
summary_df = pd.DataFrame({"Number of Unique Items":[Number_unique_items],
                              "Average Purchase Price":[avg_price],  
                              "Total Number of Purchases": [number_purchases],
                              "Total Revenue": [total_revenue]})
summary_df.head()


summary_df.style.format({"Average Purchase Price":"${:,.2f}",
                                    "Total Number of Purchases":"{:,.2f}",
                                     "Total Revenue":"${:,.2f}"})
```

## Gender Demographics

* Percentage and Count of Male Players


* Percentage and Count of Female Players


* Percentage and Count of Other / Non-Disclosed




```python
#Eliminate Duplicate
Dropped_df = purchase_data.drop_duplicates(
        subset=["SN"])


sr_count = Dropped_df.groupby(["Gender"])["SN"].count()
sr_count
total_player = Player_Count_df.iloc[0,0]

Gen_percents= (sr_count/ total_player)*100
Gen_percents



Summary_table=pd.DataFrame({"Total Count":sr_count,
                              "Percent of Players":Gen_percents})

Summary_table.sort_values(by="Total Count", ascending=False)


Summary_table.style.format({"Percent of Players":"{:,.2f}%"})
```


## Purchasing Analysis (Gender)

* Run basic calculations to obtain purchase count, avg. purchase price, avg. purchase total per person etc. by gender




* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame

```python
gender_count=purchase_data.groupby(["Gender"]).count()["Price"].rename("Purchase Count")

gender_avg=purchase_data.groupby(["Gender"]).mean()["Price"].rename("Average Purchase Price")

gender_total_purchase =purchase_data.groupby(["Gender"]).sum()["Price"].rename("Total Purchase Value")

gender_avg_purchase= (gender_total_purchase/Summary_table["Total Count"])

purchasing_analysis_df=pd.DataFrame({"Purchase Count" : gender_count , "Average Purchase Price" : gender_avg , 
                                     "Total Purchase": gender_total_purchase , "Avg Total Purchase per Person" : gender_avg_purchase})
purchasing_analysis_df

#Format
purchasing_analysis_df.style.format({"Average Purchase Price":"${:,.2f}",
                                     "Total Purchase":"${:,.2f}",
                                     "Avg Total Purchase per Person":"${:,.2f}"})
```

## Age Demographics

* Establish bins for ages


* Categorize the existing players using the age bins. Hint: use pd.cut()


* Calculate the numbers and percentages by age group


* Create a summary data frame to hold the results


* Optional: round the percentage column to two decimal points


* Display Age Demographics Table


```python
#Age Demographics
age_bins = [0, 9.90, 14.90, 19.90, 24.90, 29.90, 34.90, 39.90, 99999]
age_titles = ["<10", "10-14", "15-19", "20-24", "25-29", "30-34", "35-39", "40+"]

Dropped_df["Age Ranges"] = pd.cut(Dropped_df["Age"],age_bins, labels=age_titles)

age_count=Dropped_df["Age Ranges"].value_counts()

age_percent= (age_count/Players)*100

age_count_df=pd.DataFrame({"Total Count" : age_count, "Percentage of Players" : age_percent})

age_count_df=age_count_df.sort_index()
age_count_df

# Format
age_count_df.style.format({"Percentage of Players":"{:,.2f}%"})
```

## Purchasing Analysis (Age)

* Bin the purchase_data data frame by age


* Run basic calculations to obtain purchase count, avg. purchase price, avg. purchase total per person etc. in the table below


* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame

```python
age_bins = [0, 9.90, 14.90, 19.90, 24.90, 29.90, 34.90, 39.90, 99999]
age_titles = ["<10", "10-14", "15-19", "20-24", "25-29", "30-34", "35-39", "40+"]

purchase_data["Age Ranges"] = pd.cut(purchase_data["Age"],age_bins, labels=age_titles)

age_count=purchase_data.groupby(["Age Ranges"]).count()["Price"].rename("Purchase Count")

avg_age=purchase_data.groupby(["Age Ranges"]).mean()["Price"].rename("Average Purchase Price")

age_total=purchase_data.groupby(["Age Ranges"]).sum()["Price"].rename("Total Purchase Value")


age_avg_purchase= (age_total/age_count_df["Total Count"])

purchasing_analysis_age_df=pd.DataFrame({"Purchase Count" : age_count , "Average Purchase Price" : avg_age , 
                                     "Total Purchase": age_total , "Avg Total Purchase per Person" : age_avg_purchase})
purchasing_analysis_age_df

purchasing_analysis_age_df.style.format({"Average Purchase Price":"${:,.2f}",
                                     "Total Purchase":"${:,.2f}",
                                     "Avg Total Purchase per Person":"${:,.2f}"})
```

## Most Popular Items

* Retrieve the Item ID, Item Name, and Item Price columns


* Group by Item ID and Item Name. Perform calculations to obtain purchase count, average item price, and total purchase value


* Create a summary data frame to hold the results

```python
popular_items = purchase_data.loc[:,["Item ID", "Item Name", "Price"]]

total_item_purchase = popular_items.groupby(["Item ID","Item Name"]).sum()["Price"].rename("Total Purchase Value")

average_item_purchase = popular_items.groupby(["Item ID","Item Name"]).mean()["Price"]

item_count = popular_items.groupby(["Item ID","Item Name"]).count()["Price"].rename("Purchase Count")

# displaying table
popular_items_summary = pd.DataFrame({"Total Purchase Value": total_item_purchase, 
                                      "Item Price": average_item_purchase, 
                                      "Purchase Count": item_count})

#popular_items_summary = popular_items_summary.loc[:,["Purchase Count", "Item Price", "Total Purchase Value"]]


```
